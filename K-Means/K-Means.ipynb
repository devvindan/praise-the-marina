```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
from scipy.spatial.distance import cdist
from sklearn.datasets import load_sample_image
from scipy import misc
from matplotlib.image import imread
```

```python
n_clusters = 5
```

```python
X, labels = make_blobs(n_samples = 1000, n_features=2, centers=n_clusters, cluster_std=0.5, random_state=20)
```

```python
plt.scatter(X[:, 0], X[:, 1], c='gray')
```

```python
class OurKMeans():

    def __init__(self, n_clusters=5, max_iter=300, tolerance=1e-4, seed=2019):

        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.tolerance = tolerance
        self.seed = seed
        self.cluster_centers_ = None
        self.labels_ = None
        
    def fit(self, X):
        
        
        np.random.seed(seed=self.seed)
        random_indicies = np.random.choice(list(range(len(X))), size=self.n_clusters, replace=False)
        self.cluster_centers_ = np.array(X[random_indicies, :])
        self.labels_ = np.zeros((len(X), 1))
        
        for iteration in range(self.max_iter):
            
            # compute distances to each centroid
            distances = cdist(X, self.cluster_centers_)
            
            self.labels_ = np.argmin(distances, axis=1)
            new_centroids = np.array([np.mean(X[self.labels_==i,:], axis=0) for i in range(self.n_clusters)])
            
            if np.linalg.norm(new_centroids - self.cluster_centers_) < self.tolerance:       
                print("Converged at iteration: {}".format(iteration))
                self.cluster_centers_ = new_centroids
                return
            
            self.cluster_centers_ = new_centroids
```

```python
our_estimator = OurKMeans(n_clusters=n_clusters, seed=5)
lib_estimator = KMeans(n_clusters=n_clusters)
```

```python
our_estimator.fit(X)
```

```python
our_labels = our_estimator.labels_
good_labels = lib_estimator.fit_predict(X)
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1,3,1)
plt.title("True labels")
plt.scatter(X[:, 0], X[:, 1], c=labels)

plt.subplot(1,3,2)
plt.title("Our labels")
plt.scatter(X[:, 0], X[:, 1], c=our_labels)
plt.scatter(our_estimator.cluster_centers_[:, 0],
           our_estimator.cluster_centers_[:, 1], c='red', marker='x')

plt.subplot(1,3,3)
plt.title("SKLearn K-means")
plt.scatter(X[:, 0], X[:, 1], c=good_labels)
plt.scatter(lib_estimator.cluster_centers_[:, 0],
           lib_estimator.cluster_centers_[:, 1], c='red', marker='x')
```

Use of K-Means to select top-N colors in image:

```python
image = imread('moon.jpg')
image = np.array(image, dtype=np.float64) / 255
plt.figure(figsize = (20,10))
plt.imshow(image)
```

```python
w, h, c = image.shape
```

```python
image_flat = image.reshape((w * h, c))
```

```python
kmeans = KMeans(n_clusters=8).fit(image_flat)
```

```python
pixel_labels = kmeans.labels_
```

```python
simplified_image = np.array([kmeans.cluster_centers_[label] for label in pixel_labels]).reshape((w, h, c))
```

```python
plt.figure(figsize = (20,10))
plt.imshow(simplified_image)
```
