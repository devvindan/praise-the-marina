```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
n_clusters = 2
```

```python
X, labels = make_blobs(n_samples = 10, n_features=2, centers=n_clusters, cluster_std=0.2, center_box=(-5, 5))
```

```python
plt.scatter(X[:, 0], X[:, 1], c='gray')
```

## Критерий минимума

```python
Z = linkage(X, 'single')
```

```python
fig = plt.figure(figsize=(10, 15))
dn = dendrogram(Z)
plt.show()
```

## Критерий максимума

```python
Z = linkage(X, 'complete')
```

```python
fig = plt.figure(figsize=(10, 15))
dn = dendrogram(Z)
plt.show()
```
