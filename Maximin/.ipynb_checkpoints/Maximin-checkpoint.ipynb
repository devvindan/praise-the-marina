```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
from scipy.spatial.distance import cdist
from sklearn.datasets import load_sample_image
from scipy import misc
from matplotlib.image import imread
```

```python
n_clusters = 8
```

```python
X, labels = make_blobs(n_samples = 2400, n_features=2, centers=n_clusters, cluster_std=0.5, random_state=20)
```

```python
plt.scatter(X[:, 0], X[:, 1], c='gray')
```

```python
class Maximin():
    
    def __init__(self, seed=2019):
        
        self.seed = seed
        
    def fit(self, X):
        
        np.random.seed(self.seed)
        
        random_index = np.random.randint(0, len(X))
        
        self.n_classes = 0
        initial_cluster = X[random_index, :].reshape(1, -1)
        clusters = initial_cluster

        X = np.delete(X, random_index, axis=0)
   
        distances = cdist(initial_cluster, X)
        
        second_cluster_index = np.argmax(distances)
        second_cluster = X[second_cluster_index]
        clusters = np.vstack((clusters, second_cluster))
        X = np.delete(X, second_cluster_index, axis=0)

        while True:

            distances = cdist(X, clusters)
            labels = np.argmin(distances, axis=1)

            candidates_distances = np.array([])
            candidates = []
            
            for i, cluster in enumerate(clusters):
                
                cluster_X = X[labels == i, :]
                
                dist = cdist(cluster.reshape(1,-1), cluster_X)
                candidates_distances = np.append(candidates_distances, np.max(dist))
                candidates.append(cluster_X[np.argmax(dist), :])
                
            top_candidate_distance = np.max(candidates_distances)
            top_candidate = np.array(candidates[np.argmax(candidates_distances)])
            
            intracluster_mean_distances = []
            
            for i in range(len(clusters)):
                for j in range(i, len(clusters)):
                    intracluster_mean_distances.append(np.linalg.norm(clusters[i] - clusters[j]))
                    
            mean_distance = np.mean(intracluster_mean_distances)
            
            if (top_candidate_distance > mean_distance / 2.0):
                clusters = np.vstack((clusters, top_candidate))
                top_index = np.argwhere(X == top_candidate)[0][0]
                X = np.delete(X, top_index, axis=0)           
            else:   
                return clusters       
```

```python
estimator = Maximin()
```

```python
clusters = estimator.fit(X)
```

Algorithm returned only 7 clusters

```python
clusters
```

```python
distances = cdist(X, clusters)
our_labels = np.argmin(distances, axis=1)
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1,2,1)
plt.title("True labels")
plt.scatter(X[:, 0], X[:, 1], c=labels)

plt.subplot(1,2,2)
plt.title("Our labels")
plt.scatter(X[:, 0], X[:, 1], c=our_labels)
plt.scatter(clusters[:, 0], clusters[:, 1], color='red', marker='x')
```

Initialize K-Means using cluster centroids

```python
k_means_ours = KMeans(n_clusters=clusters.shape[0], init=clusters, n_init=1)
k_means_default = KMeans(n_clusters=clusters.shape[0])
```

```python
our_k_labels = k_means_ours.fit_predict(X)
default_k_labels = k_means_default.fit_predict(X)
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2,2,1)
plt.title("True labels")
plt.scatter(X[:, 0], X[:, 1], c=labels)

plt.subplot(2,2,2)
plt.title("Our labels")
plt.scatter(X[:, 0], X[:, 1], c=our_labels)
plt.scatter(clusters[:, 0], clusters[:, 1], color='red', marker='x')

plt.subplot(2,2,3)
plt.title("K-means labels with our initialization")
plt.scatter(X[:, 0], X[:, 1], c=our_k_labels)
plt.scatter(k_means_ours.cluster_centers_[:, 0],
            k_means_ours.cluster_centers_[:, 1], c='red', marker='x')

plt.subplot(2,2,4)
plt.title("K-means labels with default initialization")
plt.scatter(X[:, 0], X[:, 1], c=default_k_labels)
plt.scatter(k_means_default.cluster_centers_[:, 0],
            k_means_default.cluster_centers_[:, 1], c='red', marker='x')
```
