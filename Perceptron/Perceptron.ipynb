```python
from sklearn.datasets import make_classification
from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt
import numpy as np
```

```python
classifier = Perceptron(alpha=0.0001, max_iter=300, tol=None)
```

```python
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                           random_state=1, n_clusters_per_class=1)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
# сейчас бы руками персептрон писать

classifier.fit(X, y)
```

```python
xx, yy = np.mgrid[-2:1.5:.01, 0:2.5:.01]
grid = np.c_[xx.ravel(), yy.ravel()]
probs = classifier.decision_function(grid).reshape(xx.shape)
```

```python
plt.contour(xx, yy, probs, levels=[.5], cmap="Greys", vmin=0, vmax=.6)
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
our_point = [[1.5, 1.0]]
```

```python
plt.contour(xx, yy, probs, levels=[.5], cmap="Greys", vmin=0, vmax=.6)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.scatter(our_point[0][0], our_point[0][1], c='green')
```

```python
prediction = classifier.predict(our_point)[0]
```

```python
if prediction == 1:
    print("Yellow")
else:
    print("Violet")
```
